

## Задание

Дана функция f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30

1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
from sympy import *
from sympy.plotting import plot
from sympy.solvers import solve
from sympy import Symbol
```

## Вид функции


```python
x = Symbol('x')
y = -12 * (x**4) * sin(cos(x)) - 18 * (x**3) + 5 * (x**2) + 10*x - 30
y
```

## График функции на промежутке (-10; 10)

```python
graph = plot(y, (x,-10, 10))
```

## Корни уравнения

```python
r = nonlinsolve([y], x)
print(r)
```

## Интервалы возрастания функции

```python
solveset(y>0)
```

## Интервалы убывания функции

```python
solveset(y<0)
```

## Точки экстремума функции

```python
pr_expr = y.diff(x)
nonlinsolve([pr_expr],[x])
```

## Функция принимает положительные значения:

```python
solveset(y>0)
```

## И отрицательные значения:

```python
solveset(y<0)
```
